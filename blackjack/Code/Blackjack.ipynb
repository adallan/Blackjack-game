```python
#Global variables

import random
suits =('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':10, 'Queen':10, 'King':10, 'Ace':1}
```

```python
class Card():
    
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
        
    def __str__(self):
        return self.rank + ' of ' + self.suit
```

```python
class Deck():
    
    def __init__ (self):
        
        self.all_cards = []
        
        for suit in suits:
            for rank in ranks:
                #create card object
                created_card = Card(suit,rank)
                
                self.all_cards.append(created_card)
                
    def shuffle(self):
        
        random.shuffle(self.all_cards)
        
    def deal_one(self):
        return self.all_cards.pop(0)
```

```python
class Player():
    def __init__(self,owner,balance,score,hand):
        
        self.owner = owner
        self.balance = balance
        self.score = score
        self.hand = hand
        
    def deposit(self,amount):
        
        self.balance = self.balance + amount
        print('Available betting pool: ')
        return self.balance
        
    def place_bet(self,amount):
        
        if amount <= self.balance:
            self.balance = self.balance - amount
            print('Bet placed ')
            return self.balance
        else:
            print('Funds unavaible')
            
    def new_score(self,card):
        self.score = self.score + card.value
        
    def add_card(self,new_card):
        self.hand.append(new_card)
        
    def check_ace(self):
        found_ace = False
        
        for card in self.hand:
            if card.rank == 'Ace':
                found_ace = True
                
        if found_ace and self.score  <= 11:
            self.score += 10
            
    def __str__(self):
        return f'Available betting balance: {self.balance} '
```

```python
def replay():
    
    answer = 'Invalid'
    yes = 'Y'
    no = 'N'
    
    while answer == 'Invalid':
        replay_choice = input('Do you want to play again? (Y/N): ')
        if replay_choice.upper() in yes:
            return True
        if replay_choice.upper() in no:
            return False
```
